# Fake News Detection using LSTMs

### Dataset  - https://www.kaggle.com/datasets/saurabhshahane/fake-news-classification 

```python
import numpy as np
import pandas as pd


import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer

from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

### About dataset: 
1. Serial number starting from 0
2. Title about the text news heading
3. Text about the news content
4. Label [0: fake and 1: real]

```python
df = pd.read_csv("data/welfake_dataset.csv", index_col=0)
df.head()
```

```python
df.shape
```

## Data Preprocessing

```python
# Drop missing values
df = df.dropna()
```

```python
df.shape
```

```python
# Dependent and Independent features
X = df.drop("label", axis = 1)
y = df["label"]

X.shape, y.shape
```

```python
X = X.reset_index()
```

```python
# Vocabulary Size
vocab_size = 5000
```

```python
# Cleaning the text data in the dataset
stemmer = PorterStemmer()
corpus = []

for i in range(0, len(X)):
    print("At Step: {}".format(i))
    
    review = re.sub('[^a-zA-Z]', ' ', X["title"][i])
    review = review.lower()
    review = review.split()
    
    review = [stemmer.stem(word) for word in review if not word in stopwords.words("english")]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Creating One-hot representations
onehot_representation = [one_hot(words, vocab_size) for words in corpus]
onehot_representation
```

```python
find_max = []
for sentences in corpus:
    find_max.append(len(sentences.split()))
```

```python
# Finding the length of sentences in corpus at 99th percentile 
np.percentile(find_max, 99)
```

## Word Embedding - Feature Representation

```python
sentence_length = 20
embedded_docs = pad_sequences(onehot_representation, 
                              padding = "pre", 
                              maxlen = sentence_length)

embedded_docs
```

```python
len(embedded_docs)
```

## Creating Model

```python
embedding_vector_features = 80

model = Sequential()
model.add(Embedding(vocab_size, embedding_vector_features, input_length = sentence_length))
model.add(LSTM(100))
model.add(Dense(1, activation = "sigmoid"))
```

```python
model.compile(
        optimizer = "adam", 
        loss = "binary_crossentropy", 
        metrics = ["accuracy"]
)
```

```python
model.summary()
```

```python
len(embedded_docs), y.shape
```

```python
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape, y_final.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, 
                                                    test_size = 0.33, 
                                                    random_state = 40)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Training the Model

```python
model.fit(X_train, y_train, 
          validation_data = (X_test, y_test), 
         epochs = 30, 
         batch_size = 64)
```

## Performance Metrics 

```python
y_preds = model.predict(X_test)
y_preds
```

```python
loss, acc = model.evaluate(X_test, y_test)
```

## Adding a dropout layer to boost performance

```python
embedding_vector_features = 80

model_2 = Sequential()
model_2.add(Embedding(vocab_size, embedding_vector_features, input_length = sentence_length))
model_2.add(tf.keras.layers.Dropout(0.3))
model_2.add(LSTM(100))
model_2.add(tf.keras.layers.Dropout(0.3))
model_2.add(Dense(1, activation = "sigmoid"))
```

```python
model_2.compile(
        optimizer = "adam", 
        loss = "binary_crossentropy", 
        metrics = ["accuracy"]
)
```

```python
model_2.summary()
```

```python
# Creating a checkpoint for model
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint("Best_model_using_dropout.h5", save_best_only = True)
```

```python
model_2.fit(X_train, y_train, 
          validation_data = (X_test, y_test), 
         epochs = 30, 
         batch_size = 64, 
         callbacks = [checkpoint_cb])
```

```python
tf.keras.backend.clear_session()
```

```python
loss, acc = model_2.evaluate(X_test, y_test)
```
